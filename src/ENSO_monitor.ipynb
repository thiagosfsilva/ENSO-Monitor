```python
import requests
import pandas as pd
import numpy as np
```

```python
ana_url = 'http://telemetriaws1.ana.gov.br/ServiceANA.asmx/DadosHidrometeorologicos'
codEstacao = '12850000'
dataInicio = '2023-01-01'
dataFim = '2023-12-01'
```

```python
response = requests.get(ana_url, params = {"codEstacao": codEstacao, "dataInicio": dataInicio, "dataFim": dataFim })
```

```python
rawTable = pd.read_xml(response.content, namespaces={'msdata':"urn:schemas-microsoft-com:xml-msdata"}, xpath='.//DadosHidrometereologicos')
rawTable.head()
```

```python
rawTable.dropna()
rawTable['DataHora'] = pd.to_datetime(rawTable['DataHora'])
rawTable['Dt'] = pd.to_datetime(rawTable['DataHora']).dt.date
rawTable['Hr'] = pd.to_datetime(rawTable['DataHora']).dt.time
rawTable.head()
```
